## Three observable trends based on the data.

* The Tumor Response to Treatment shows that mice that were treated through the Capomulin regime show descrease in tumor volue by 11.2 mm3 over the course of tratment, while mice that were getting  Infubinol, Ketapril and Placebo treatments showed increase in tumor volue by 20-25 mm3
* All mice showed metatatic spread during course of a treatment, however the mice that were treated by Capomulin showed the lowest increase in metastatic sites (1.476190). Mice that were treated with Ketapril showed the greatest increase (3.363636)


```python
##### Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import matplotlib.ticker as mtick

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_drug_data = pd.read_csv(mouse_drug_data_to_load)
clinical_trial_data = pd.read_csv(clinical_trial_data_to_load)

# # Combine the data into a single dataset
combined_clinical_data = pd.merge(clinical_trial_data, mouse_drug_data, on="Mouse ID")

# Display the data table for preview
combined_clinical_data.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
meanTumorVolume = combined_clinical_data.groupby(['Drug', 'Timepoint']).mean()['Tumor Volume (mm3)']\
                .reset_index()

# Preview DataFrame
meanTumorVolume.head(15)
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
se_tumorVolumes = combined_clinical_data.groupby(['Drug', 'Timepoint']).sem()['Tumor Volume (mm3)'].reset_index()

#Extracting Standard Error Values for the drug of interest
se_capomulin = se_tumorVolumes[se_tumorVolumes["Drug"]=="Capomulin"]["Tumor Volume (mm3)"]
se_infubinol = se_tumorVolumes[se_tumorVolumes["Drug"]=="Infubinol"]["Tumor Volume (mm3)"]
se_ketapril = se_tumorVolumes[se_tumorVolumes["Drug"]=="Ketapril"]["Tumor Volume (mm3)"]
se_placebo = se_tumorVolumes[se_tumorVolumes["Drug"]=="Placebo"]["Tumor Volume (mm3)"]


# Preview DataFrame
se_tumorVolumes.head(5)
```

```python
# Minor Data Munging to Re-Format the Data Frames
formatted_meanTumVol = meanTumorVolume.pivot(index='Timepoint', columns='Drug', values='Tumor Volume (mm3)')\
            .reset_index()

# Preview that Reformatting worked
formatted_meanTumVol.head()
```

```python
# Generate the Plot (with Error Bars)
# Generate Title, x and y labels
plt.title("Tumor Response to Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Tumor Volume (mm3)")

# Set the upper and lower limits of y axis
plt.ylim(32, 80)

# Set the upper and lower limits of x axis
plt.xlim(-2,50)

#Create the time variable to use in the chart
time = formatted_meanTumVol["Timepoint"]

# Create individual scatter plots with error bars for each drug

capomulin = plt.errorbar(time, formatted_meanTumVol["Capomulin"], yerr=se_capomulin,\
                         color="Red", linewidth=0.5, linestyle="dashed", label="Capomulin", marker="o")

infubinol = plt.errorbar(time, formatted_meanTumVol["Infubinol"], yerr=se_infubinol,\
                         color="Blue", linewidth=0.5, linestyle="dashed", label="Infubinol", marker="^")

ketapril = plt.errorbar(time, formatted_meanTumVol["Ketapril"], yerr=se_ketapril,\
                        color="Green", linewidth=0.5, linestyle="dashed", label="Ketapril", marker="s")

placebo = plt.errorbar(time, formatted_meanTumVol["Placebo"], yerr=se_placebo,\
                       color="Black", linewidth=0.5, linestyle="dashed", label="Placebo", marker="d")

# Customize the grid
plt.grid(linestyle='-', linewidth='0.5', color='grey')

# Create the legend
plt.legend(handles=[capomulin, infubinol, ketapril, placebo])

# Save the Figure
plt.savefig("images/TumorResponseToTreatment.png")

# Show the Figure
plt.show()

```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
mean_MetSites = combined_clinical_data.groupby(['Drug', 'Timepoint']).mean()['Metastatic Sites']\
                 .reset_index()
# Preview DataFrame
mean_MetSites.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
se_MetSites = combined_clinical_data.groupby(['Drug', 'Timepoint']).sem()['Metastatic Sites']\
            .reset_index()

#Extracting Standard Error Values for the drug of interest
se_capomulinResponse = se_MetSites[se_MetSites["Drug"]=="Capomulin"]["Metastatic Sites"]
se_infubinolResponse = se_MetSites[se_MetSites["Drug"]=="Infubinol"]["Metastatic Sites"]
se_ketaprilResponse = se_MetSites[se_MetSites["Drug"]=="Ketapril"]["Metastatic Sites"]
se_placeboResponse = se_MetSites[se_MetSites["Drug"]=="Placebo"]["Metastatic Sites"]

# Preview DataFrame
se_MetSites.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
formatted_mean_MetSites = mean_MetSites.pivot(index='Timepoint', columns='Drug', values='Metastatic Sites')\
            .reset_index()

# Preview that Reformatting worked
formatted_mean_MetSites.head(15)
```

```python
# Generate the Plot (with Error Bars)
# Generate Title, x and y labels
plt.title("Metatatic Spread During Treatment")
plt.xlabel("Treatment Duration (Days)")
plt.ylabel("Met. Sites")

#Create the duration variable to use in the chart
duration = formatted_mean_MetSites["Timepoint"]

# Set the upper and lower limits of y axis
plt.ylim(-0.3, 4)

# Set the upper and lower limits of x axis
plt.xlim(-3,50)

# Create individual scatter plots with error bars for each drug
capomulinResponse = plt.errorbar(duration, formatted_mean_MetSites["Capomulin"], yerr=se_capomulinResponse,\
                         color="Red", linewidth=0.5, linestyle="dashed", label="Capomulin", marker="o")

infubinolResponse = plt.errorbar(duration, formatted_mean_MetSites["Infubinol"], yerr=se_infubinolResponse,\
                         color="Blue", linewidth=0.5, linestyle="dashed", label="Infubinol", marker="^")

ketaprilResponse = plt.errorbar(duration, formatted_mean_MetSites["Ketapril"], yerr=se_ketaprilResponse,\
                        color="Green", linewidth=0.5, linestyle="dashed", label="Ketapril", marker="s")

placeboResponse = plt.errorbar(duration, formatted_mean_MetSites["Placebo"], yerr=se_placeboResponse,\
                       color="Black", linewidth=0.5, linestyle="dashed", label="Placebo", marker="d")


# Create the legend
plt.legend(handles=[capomulinResponse, infubinolResponse, ketaprilResponse, placeboResponse])

# Customize the grid
plt.grid(linestyle='-', linewidth='0.5', color='grey')

# Save the Figure
plt.savefig("images/MetataticSpreadDuringTreatment.png")

# Show the Figure
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
grouped_MouseCount = combined_clinical_data.groupby(["Drug", "Timepoint"]).count()["Mouse ID"]\
            .reset_index()

# Preview DataFrame
grouped_MouseCount.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
formatted_MouseCount = grouped_MouseCount.pivot(index='Timepoint', columns='Drug', values='Mouse ID')

# Preview the Data Frame
formatted_MouseCount
```

```python
# Generate the Plot (Accounting for percentages)
# Generate Title, x and y labels
plt.title("Survival During Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Survival Rate (%)")

#Calculating survival percent values for "Capomulin", "Infubinol", "Ketapril","Placebo"
reduced_MouseCount = formatted_MouseCount[["Capomulin", "Infubinol", "Ketapril","Placebo"]]/25*100

# Create individual plots for each drug
capomulinSurvival = plt.plot(reduced_MouseCount.index, reduced_MouseCount["Capomulin"],\
                         color="Red", linewidth=0.5, linestyle="dashed", label="Capomulin", marker="o")

infubinolSurvival = plt.plot(reduced_MouseCount.index, reduced_MouseCount["Infubinol"],\
                         color="Blue", linewidth=0.5, linestyle="dashed", label="Infubinol", marker="^")

ketaprilSurvival = plt.plot(reduced_MouseCount.index, reduced_MouseCount["Ketapril"], \
                        color="Green", linewidth=0.5, linestyle="dashed", label="Ketapril", marker="s")

placeboSurvival = plt.plot(reduced_MouseCount.index, reduced_MouseCount["Placebo"],\
                       color="Black", linewidth=0.5, linestyle="dashed", label="Placebo", marker="d")

# Create legend
plt.legend(["Capomulin", "Infubinol", "Ketapril", "Placebo"])

# Customize the grid
plt.grid(linestyle='-', linewidth='0.5', color='grey')

# Save the Figure
plt.savefig("images/SurvivalDuringTreatment.png")

# Show the Figure
plt.show()
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
formatted_TumVol = meanTumorVolume.pivot(index='Timepoint', columns='Drug', values='Tumor Volume (mm3)')\
            .reset_index()

# Extract data from the first and last rows
formatted_TumVol_firstRow = formatted_TumVol.iloc[0,1:]
formatted_TumVol_lastRow = formatted_TumVol.iloc[-1,1:]

# Calculate percent tumor volume change for each drug across the full 45 days
percent = (formatted_TumVol_lastRow - formatted_TumVol_firstRow) / formatted_TumVol_firstRow *100

# Display the data to confirm
percent
```

```python
# Store all Relevant Percent Changes into a Tuple
thistuple = (percent["Capomulin"], percent["Infubinol"], percent["Ketapril"], percent["Placebo"])

# Create values for y axis
y_ticks = [-20, 0, 20, 40, 60]

# Create values for x axis
x_axis = ["Capomulin", "Infubinol", "Ketapril", "Placebo"]

# Splice the data between passing and failing drugs
neg_values = plt.bar(x_axis[0], thistuple[0], width=1, color='green', align = "edge")
pos_values = plt.bar(x_axis[1:], thistuple[1:], width=1, color='red',  align = "edge")

# Create title, grid and name y label and 
plt.yticks(y_ticks)
plt.title("Tumor Change Over 45 Day Treatment")
plt.ylabel("% Tumor Volume Change")
plt.grid(True)
       
# Use functions to label the percentages of changes
def autolabel_neg(bars):
   for bar in bars:
       y_value = bar.get_height()
       plt.text(bar.get_x() + bar.get_width()/2., -10,
              '%d%%' % int(y_value), ha='center', va='bottom', color="white")
    
def autolabel_pos(bars):
   for bar in bars:
       y_value = bar.get_height()
       plt.text(bar.get_x() + bar.get_width()/2., 10,
             '%d%%' % int(y_value), ha='center', va='bottom', color="white")
    

# Call functions to implement the function calls
autolabel_neg(neg_values)
autolabel_pos(pos_values)  

# Save the Figure
plt.savefig("images/TumorChangeOver45DayTreatment.png")

# Show the Figure
plt.show()
```
